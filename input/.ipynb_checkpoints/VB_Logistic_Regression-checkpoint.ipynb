# ベイジアンロジスティック回帰の実装
* 参考  
ベイズ推論による機械学習入門

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit
%matplotlib inline
```

```python
def lr_create_sample(x, c):
    w = np.random.multivariate_normal(np.zeros(x.shape[1]), c)
    y = np.random.binomial(1, expit(x.dot(w)))
    return y, w
```

```python
n = 50 # the number of data points
d = 2 # input dimension
cov = np.eye(d) * 100. # prior on weight
x = 2 * np.random.rand(n, d) - 1. # sample features
y, w = lr_create_sample(x, cov) # get sample targets and weights
print("w: ", w)
print("cov: ", cov)
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
def calc_sigma(rho):
    return np.log(1 + np.exp(rho))

def softmax(a):
    exp_a = np.exp(a - np.max(a))
    return exp_a / np.sum(exp_a)
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
#     return np.exp(a) / np.sum(np.exp(a))

# def softmax(x):
#     exp_x = np.exp(x - x.max(axis=1, keepdims=True))
#     y = exp_x / np.sum(exp_x, axis=1, keepdims=1)
#     return y


class vi_for_logistic_regression:
    def __init__(self, x, y, lr=0.0001, iter_num=10000):
        self.x = x
        self.y = y
        self.sigma_w_prior = 100 * np.eye(x.shape[1])# prior sigma for w
        self.lr = lr
        self.iter_num = iter_num
    def calc_dg_dw(self, x, y, sigma_w_prior, mu, rho, w):
        dq_dw = (mu - w) / (calc_sigma(rho) ** 2)
        dp_dw = np.diag(sigma_w_prior) * w
#         return dq_dw + dp_dw - x.T.dot(y - softmax(x.dot(w)))
        return dq_dw + dp_dw - x.T.dot(y - sigmoid(x.dot(w)))
    def calc_dg_dmu(self, w, mu, rho):
        return (w - mu) / (calc_sigma(rho) ** 2)
    def calc_dg_drho(self, w, mu, rho):
        return - (1 / 2) * ((w - mu) ** 2 - calc_sigma(rho) ** 2) * self.calc_d_drho(rho)
    def calc_d_drho(self, rho):
        return 2 * calc_sigma(rho) ** (-3) * (1 / (1 + np.exp(rho))) ** 2 * (1 / (1 + np.exp(-rho)))
    def fit(self):
        self.mu_list = []
        self.rho_list = []
        self.mu = np.random.normal(size=x.shape[1])
        self.rho = np.random.normal(size=x.shape[1])
        for i in range(self.iter_num):
#             eps = np.random.rand()
            eps = np.random.normal(0, 1)
            w_sample = self.mu + calc_sigma(self.rho) * eps
#             print(w_sample)
            dg_dw = self.calc_dg_dw(x, y, self.sigma_w_prior, self.mu, self.rho, w_sample)
            dg_dmu = self.calc_dg_dmu(w_sample, self.mu, self.rho)
            dg_drho = self.calc_dg_drho(self.rho, self.mu, self.rho)
            
            d_mu = dg_dw + dg_dmu
            d_rho = dg_dw * (eps / (1 + np.exp(-self.rho))) + dg_drho
            
            self.mu = self.mu - self.lr * d_mu
            self.rho = self.rho - self.lr * d_rho
            self.mu_list.append(self.mu)
            self.rho_list.append(self.rho)
#             print(dg_dmu, dg_drho)
    
        return self.mu, self.rho
    
    def predict_prob(self, data):
        return 1/ (1 + np.exp(-data.dot(self.mu)))
    def predict_label(self, data):
        prob = self.predict_prob(data)
        return np.where(prob < 0.5, 0, 1)
    
```

```python
vi_lr = vi_for_logistic_regression(x, y)
```

```python
m, r = vi_lr.fit()
print("期待値: {} {}".format(vi_lr.mu[0], vi_lr.mu[1]))
print("精度: {} {}".format(vi_lr.rho[0], vi_lr.rho[1]))
```

## パラメータの事後分布の表示

```python
plt.hist(np.random.normal(m[0], calc_sigma(r[0]) ** 2, size=10000), bins=100, alpha=0.5, color="blue")
plt.hist(np.random.normal(m[1], calc_sigma(r[1]) ** 2, size=10000), bins=100, alpha=0.5, color="red")
plt.show()
```

## 分離平面の式
* 切片は$\log{\frac{1 - p}{p}}$に注意
* 確率0.5の平面を図示。事後分布からランダムに10個発生。
* 濃青線は事後分布の平均値による分離平面の式

```python
# xmin = -4
# xmax = 4
# ymin = -4
# ymax = 4
x_ = np.linspace(-4, 4)
y_ = - (m[0] / m[1]) * x_
plt.figure(figsize=(7, 7))
plt.plot(x_, y_, label='mean')
plt.scatter(x[:,0], x[:,1], c=y)
sig = calc_sigma(r)
for _ in range(10):
    w_s = np.random.multivariate_normal(m, np.diag(sig**2))
    y_r = - (w_s[0] / w_s[1]) * x_
    plt.plot(x_, y_r, alpha=0.5)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.show()
```

## 10個の事後分布からのサンプルによる各点の確率の平均値の等高線
* 事後分布からランダムに10個直線をサンプルする。
* 各直線に基づき平面内の点の分類確率を予測する。
* 10個の値の平均値を計算し、等高線プロットで描画する。
* 濃青線で事後分布の平均値による分離平面を描画。

```python
x_ = np.linspace(-4, 4, 100)
y_ = - (m[0] / m[1]) * x_
plt.figure(figsize=(8, 8))
plt.plot(x_, y_, label='mean')
plt.scatter(x[:,0], x[:,1], c=y)
p = np.array([0.2, 0.4, 0.6, 0.8])
X, Y = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
x_test = np.array([X, Y]).transpose(1, 2, 0).reshape(-1, 2)
# plt.scatter(X, Y)
z_list = []
sig = calc_sigma(r)
for _ in range(10): # サンプル回数
    w_s = np.random.multivariate_normal(m, np.diag(sig ** 2))
    z_tmp = 1 / (1 + np.exp((-(x_test.dot(w_s))))).reshape(100, 100)
    z_list.append(z_tmp)
#     for p_ in p: # 描く境界の確率
    y_ = -(w_s[0] / w_s[1]) * x_ #- np.log((1-p_)/p_)
    plt.plot(x_, y_)
z = np.mean(np.array(z_list), axis=0)
levels = np.linspace(0, 1, 5)
print(z)
plt.contour(X, Y, z)
# plt.clabel(cp, inline=True, fontsize=10)

plt.xlim(-4, 4)
plt.ylim(-4, 4)
# plt.legend()
plt.show()
```
